```python
import pandas as pd

customer_df = pd.read_csv(
    "../data/raw/customer_support/aa_dataset-tickets-multi-lang-5-2-50-version.csv"
)

retail_df = pd.read_excel(
    "../data/raw/online_retail/Online Retail.xlsx"
)
```

```python
customer_df["subject"] = customer_df["subject"].fillna(
    "No Subject"
)
```

```python
customer_df["answer"] = customer_df["answer"].fillna(
    "No Answer"
)
```

```python
tag_columns = [
    "tag_1","tag_2","tag_3","tag_4",
    "tag_5","tag_6","tag_7","tag_8"
]

customer_df[tag_columns] = (
    customer_df[tag_columns]
    .fillna("Unknown")
)
```

```python
retail_df = retail_df.dropna(
    subset=["Description"]
)
```

```python
retail_customers = retail_df.dropna(
    subset=["CustomerID"]
)
```

```python
retail_df["Revenue"] = (
    retail_df["Quantity"]
    *
    retail_df["UnitPrice"]
)
```

```python
retail_df["InvoiceDate"] = pd.to_datetime(
    retail_df["InvoiceDate"]
)
```

```python
customer_df.to_csv(
    "../data/processed/customer_support/customer_support_clean.csv",
    index=False
)

retail_df.to_csv(
    "../data/processed/online_retail/retail_clean.csv",
    index=False
)
```
